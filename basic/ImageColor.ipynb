```python
import cv2
import matplotlib.pyplot as plt 
img = r'F:\works\A-important\A-neurals\CV\images\Image.jpg'
```

```python
img_cv=cv2.imread(img)
cv2.imshow('image',img_cv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_cv)
ax.axis("off")
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_cv[:, :, 1])
ax.axis("off")
plt.show()
```

```python
fig,ax=plt.subplots(1, 3, figsize=(7,7))
ax[0].imshow(img_cv[:, :, 0], cmap='gray')
ax[1].imshow(img_cv[:, :, 1], cmap='gray')
ax[2].imshow(img_cv[:, :, 2], cmap='gray')
ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')
plt.show()
```

```python
img_resize1=cv2.resize(img_cv,None,fx=5,fy=10)
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_resize1)
ax.axis("off")
plt.show()
```

```python
img_resize1=cv2.resize(img_cv,(1024*4, 682*4))
fig,ax=plt.subplots(figsize=(7,7))
ax.imshow(img_resize1)
ax.axis("off")
plt.show()
```

```python
img_resize3=cv2.resize(img_cv,(50,50),interpolation=cv2.INTER_CUBIC)
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_resize3)
ax.axis("off")
plt.show()
```

```python
#sharpening
import numpy as np
kernel=np.array([[-1,-1,-1],
                [-1,9,-1],
                [-1,-1,-1]])
img_sharp=cv2.filter2D(img_cv,-1,kernel)
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_sharp)
ax.axis("off")
plt.show()
```

```python
import numpy as np
kernel=np.array([[-1,-1,-1],
                [-1,8,-1],
                [-1,-1,-1]])
img_edge=cv2.filter2D(img_cv,-1,kernel)
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_edge)
ax.axis("off")
plt.show()
```

```python
import numpy as np
kernel=np.array([[1,1,1],
                [1,1,1],
                [1,1,1]])
kernel=kernel/9
img_blur=cv2.filter2D(img_cv,-1,kernel)
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_blur)
ax.axis("off")
plt.show()
```

```python
img_resized = cv2.resize(img_blur, (1000, 682))
cv2.imshow("image", img_resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_gray = cv2.cvtColor(img_blur, cv2.COLOR_BGR2GRAY)
cv2.imshow("image", img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
