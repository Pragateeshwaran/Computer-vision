```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
image = plt.imread('./image/Tree.jpg')
print(image.shape)

fig, ax = plt.subplots(1, 3, figsize=(25, 5))

ax[0].imshow(image[:, :, 0], cmap='gray')
ax[0].set_title("Red Channel")

ax[1].imshow(image[:, :, 1], cmap='gray')
ax[1].set_title("Green Channel")

ax[2].imshow(image[:, :, 2], cmap='gray')
ax[2].set_title("Blue Channel")

plt.show()
```

```python
# Crop

dimension_to_crop = (60, 200)
new_image = image[0: dimension_to_crop[0], 0: dimension_to_crop[1], :]
plt.imshow(new_image)
```

```python
#resize 

resized = cv2.resize(image, (100, 200)) 
plt.imshow(resized)
```

```python
flipped_image_horizontal = cv2.flip(image, 0)
plt.imshow(flipped_image_horizontal)
```

```python
angle = 45
center = (image.shape[1] / 2, image.shape[0] / 2)
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 0.5)
rotated_image = cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]))
plt.imshow(rotated_image)
```
