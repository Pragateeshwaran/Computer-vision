```python
import cv2
import matplotlib.pyplot as plt
import PIL as image
import numpy as np
```

```python
image1 = plt.imread("./image/Tree.jpg")
image2 = plt.imread("./image/lake.jpg")
```

```python
image1 = cv2.resize(image1, (512, 512))
image2 = cv2.resize(image2, (512, 512))
```

```python
result = cv2.bitwise_and(image1, image2)
plt.imshow(result)
```

```python
result = cv2.bitwise_or(image1, image2)
plt.imshow(result)
```

```python
result = cv2.bitwise_xor(image1, image2)
plt.imshow(result)
```

```python
res = cv2.blur(image1, (10, 10))
plt.imshow(res)
```

```python
res = cv2.GaussianBlur(image1, (7, 7), sigmaX=0)
plt.imshow(res)
```

```python
res = cv2.medianBlur(image1, 15)
plt.imshow(res)
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('.\image\lake.jpg')

interpolations = {
    'INTER_NEAREST': cv2.INTER_NEAREST,
    'INTER_LINEAR': cv2.INTER_LINEAR,
    'INTER_AREA': cv2.INTER_AREA,
    'INTER_CUBIC': cv2.INTER_CUBIC,
    'INTER_LANCZOS4': cv2.INTER_LANCZOS4
}

fig, axes = plt.subplots(1, 5, figsize=(20, 5))

for idx, (name, method) in enumerate(interpolations.items()):
    resized = cv2.resize(image, (50, 50), interpolation=method)
    axes[idx].imshow(resized)
    axes[idx].set_title(name)
    axes[idx].axis('off')

plt.tight_layout()
plt.show()
```
