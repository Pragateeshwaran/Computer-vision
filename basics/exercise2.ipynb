```python
import cv2
import matplotlib.pyplot as plt
import PIL as image
import numpy as np
```

```python
image1 = plt.imread("./image/Tree.jpg")
image2 = plt.imread("./image/lake.jpg")
```

```python
image1 = cv2.resize(image1, (512, 512))
image2 = cv2.resize(image2, (512, 512))
```

```python
result = cv2.bitwise_and(image1, image2)
plt.imshow(result)
```

```python
result = cv2.bitwise_or(image1, image2)
plt.imshow(result)
```

```python
result = cv2.bitwise_xor(image1, image2)
plt.imshow(result)
```

```python
res = cv2.blur(image1, (10, 10))
plt.imshow(res)
```

```python
res = cv2.GaussianBlur(image1, (7, 7), sigmaX=0)
plt.imshow(res)
```

```python
res = cv2.medianBlur(image1, 15)
plt.imshow(res)
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('.\image\lake.jpg')

interpolations = {
    'INTER_NEAREST': cv2.INTER_NEAREST,
    'INTER_LINEAR': cv2.INTER_LINEAR,
    'INTER_AREA': cv2.INTER_AREA,
    'INTER_CUBIC': cv2.INTER_CUBIC,
    'INTER_LANCZOS4': cv2.INTER_LANCZOS4
}

fig, axes = plt.subplots(1, 5, figsize=(20, 5))

for idx, (name, method) in enumerate(interpolations.items()):
    resized = cv2.resize(image, (50, 50), interpolation=method)
    axes[idx].imshow(resized)
    axes[idx].set_title(name)
    axes[idx].axis('off')

plt.tight_layout()
plt.show()
```

```python
sobelx = cv2.Sobel(image2, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(image2, cv2.CV_64F, 0, 1, ksize=3)
sobel = cv2.magnitude(sobelx, sobely)
plt.imshow(sobel);
```

```python
gray_image = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap="gray");
```

```python
blurred_image = cv2.GaussianBlur(gray_image, (3, 3), 1)
plt.imshow(blurred_image, cmap="gray")
```

```python
edges = cv2.Canny(blurred_image, 50, 150)
plt.imshow(edges, cmap="gray");
```

```python
prewitt_x = np.array([[-1, 0, 1],
                      [-1, 0, 1],
                      [-1, 0, 1]])
prewitt_y = np.array([[ 1,  1,  1],
                      [ 0,  0,  0],
                      [-1, -1, -1]])
prewitt_x_img = cv2.filter2D(blurred_image, -1, prewitt_x)
prewitt_y_img = cv2.filter2D(blurred_image, -1, prewitt_y)
prewitt = cv2.magnitude(prewitt_x_img.astype(np.float32), prewitt_y_img.astype(np.float32))
plt.imshow(prewitt);
```

```python
roberts_x = np.array([[1, 0],
                      [0, -1]])
roberts_y = np.array([[0, 1],
                      [-1, 0]])
roberts_x_img = cv2.filter2D(blurred_image, -1, roberts_x)
roberts_y_img = cv2.filter2D(blurred_image, -1, roberts_y)
roberts = cv2.magnitude(roberts_x_img.astype(np.float32), roberts_y_img.astype(np.float32))
plt.imshow(roberts)
```

```python
laplacian = cv2.Laplacian(blurred_image, cv2.CV_64F)
laplacian = np.abs(laplacian)
plt.imshow(laplacian)
```
